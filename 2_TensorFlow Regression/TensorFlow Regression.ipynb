```python
#now let's start feature engineering process also getting rid of features that are not going to be useful to us
df.head()
```

```python
df=df.drop('id',axis=1)
```

```python
df['date']
```

```python
#this date column looks like a string. We can convert it to datetime object
df['date']=pd.to_datetime(df['date'])
```

```python
#we can extract month and year of this.
df['year']=df['date'].apply(lambda date:date.year)
df['month']=df['date'].apply(lambda date:date.month)
```

```python
df=df.drop('date',axis=1)
```

```python
df.head()
```

```python
df['zipcode'].value_counts()
```

```python
#there are 70 categories of the zipcode and getting dummies for 70 variables is probably too much, so we drop it
df=df.drop('zipcode',axis=1)
```

```python
df['yr_renovated'].value_counts()
#categorize them to renovated or not renovated. 0 here means not renovated. while we can convert all to 1 meaning renovated
#we also keep this feature as it is, because we say that the recently the house got renovated the more price it has
```

```python
df['sqft_basement'].value_counts()
#we also keep this feature as it is, because it is intuitve that a house having a basement will have higher value. 
#so we keep this feature as it is.
```

```python
#feature engineering and EDA is an extremely important part for any data science and machine learning project
```

```python
X=df.drop('price',axis=1).values
y=df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#now we perform scaling.
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
#we have 19 incoming features. So that is a good range to have 19 neurons in our layer
```

```python
model=Sequential()
#we will explore later on, is to see how we can use early stoppings in order to try to choose the correct number to train
#for, but also to choose correct number of layers to train for.

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),batch_size=128,epochs=400)
#validation_data: Data on which to evaluate the loss and any model metrics at the end of each epoch.The model will not be 
#trained on this data.
```

```python
#explore and evaluate our data
model.history.history
```

```python
pd.DataFrame(model.history.history)
#loss on training and validation data
```

```python
#now we can directly compare loss and val_loss to see if we are overfitting to the training data on our model
losses=pd.DataFrame(model.history.history)
losses.plot()
#since the validation is going down upto 400 epochs and not increasing, that is an indicator that we could continue training
#without overfitting to our training data. If val_loss line begin to spike after some of the epoch, it means we are
#overfitting to the training data because then we have a much larger loss on the validation data.
#Further, we will talk abou early stoppings.
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
predictions=model.predict(X_test)
```

```python
mean_squared_error(y_test,predictions)
```

```python
mean_squared_error(y_test,predictions)**0.5
```

```python
mean_absolute_error(y_test,predictions)
#tells average absolute error across all the predictions
```

```python
df['price'].describe()
```

```python
explained_variance_score(y_test,predictions)
#model explains around 80 percent of the variance. Which is merely ok!
```

```python
plt.figure(figsize=(12,6))
plt.scatter(y_test,predictions)
plt.plot(y_test,y_test,'r')      #this line represents the perfect prediction line
#In the grapg it looks like we are being punished by the extremely expensive houses, the outliers. We are doing good on the
#houses between 0 and 2 miliion
```
